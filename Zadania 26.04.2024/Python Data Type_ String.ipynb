```
#2 Write a Python program to count the number of characters (character frequency) in a string:
def licz_czestotliwosc_znakow(tekst):
    czestotliwosc = {}
    for znak in tekst:
        if znak in czestotliwosc:
            czestotliwosc[znak] += 1
        else:
            czestotliwosc[znak] = 1
    return czestotliwosc
tekst = "google.com"
wynik = licz_czestotliwosc_znakow(tekst)
print(wynik)
```

```
#3 Write a Python program to get a string made of the first 2 and last 2 characters of a given string. If the string length is less than 2, return the empty string instead:
def first_and_last_two_chars(string):
    if len(string) < 2:
        return ""
    result = string[:2] + string[-2:]
    return result

print(first_and_last_two_chars('Warszawa'))
print(first_and_last_two_chars('ul'))
print(first_and_last_two_chars(' w'))
```

```
#4 	Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself:
def change_char(string):
    if len(string) < 1:
        return ""
    first_char = string[0]
    modified_string = first_char + string[1:].replace(first_char, '$')

    return modified_string

print(change_char('kraków'))
```

```
#5 Write a Python program to get a single string from two given strings, separated by a space and swap the first two characters of each string.
def zamiana(string1, string2):
    return string2[:2] + string1[2:] + ' ' + string1[:2] + string2[2:]

print(zamiana('abc', 'xyz'))
```

```
#6 Write a Python program to add 'ing' at the end of a given string (length should be at least 3). If the given string already ends with 'ing', add 'ly' instead. If the string length of the given string is less than 3, leave it unchanged.
def dodaj_koncowke(string):
    if len(string) < 3:
        return string
    elif string.endswith('ing'):
        return string + 'ly'
    else:
        return string + 'ing'

print(dodaj_koncowke('abc'))
print(dodaj_koncowke('string'))
```

```
#7 Write a Python program to find the first appearance of the substrings 'not' and 'poor' in a given string. If 'not' follows 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string.
def replace_not_poor(string):
    index_not = string.find('not')
    index_poor = string.find('poor')

    if index_not != -1 and index_poor != -1 and index_not < index_poor:
        return string[:index_not] + 'good' + string[index_poor + 4:]
    else:
        return string

print(replace_not_poor('The lyrics is not that poor!'))
print(replace_not_poor('The lyrics is poor!'))
```

```
#8 Write a Python function that takes a list of words and return the longest word and the length of the longest one
def longest_word_and_length(words):
    if not words:
        return None, 0

    longest_word = max(words, key=len)
    return longest_word, len(longest_word)

# Przykład użycia funkcji
words = ["Python", "programming", "is", "fun"]
longest, length = longest_word_and_length(words)
print("Longest word:", longest)
print("Length of the longest word:", length)
```

```
#9 Write a Python program to remove the nth index character from a nonempty string.
def remove_nth_character(string, n):
    if not string:
        return string

    if n < 0 or n >= len(string):
        return string

    return string[:n] + string[n+1:]

# Przykład użycia funkcji
original_string = "Python"
n = 2
result = remove_nth_character(original_string, n)
print("Original string:", original_string)
print("String after removing character at index", n, ":", result)
```

```
#10 Write a Python program to change a given string to a newly string where the first and last chars have been exchanged.
def exchange_first_last(string):
    if len(string) <= 1:
        return string
    else:
        return string[-1] + string[1:-1] + string[0]

# Przykład użycia funkcji
original_string = "Python"
result = exchange_first_last(original_string)
print("Original string:", original_string)
print("String after exchanging first and last characters:", result)
```

```
#11 Write a Python program to remove characters that have odd index values in a given string.
def remove_odd_index_chars(string):
    return string[::2]

# Przykład użycia funkcji
original_string = "Python"
result = remove_odd_index_chars(original_string)
print("Original string:", original_string)
print("String after removing characters with odd index values:", result)
```

```
#12 Write a Python program to count the occurrences of each word in a given sentence.
def count_word_occurrences(sentence):
    words = sentence.split()
    word_count = {}

    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count

# Przykład użycia funkcji
sentence = "This is a sample sentence. This sentence contains sample words."
word_occurrences = count_word_occurrences(sentence)
print("Word occurrences:", word_occurrences)
```

```
#14 Write a Python program that accepts a comma-separated sequence of words as input and prints the distinct words in sorted form (alphanumerically).
input_sequence = input()
words = input_sequence.split(',')
words = [word.strip() for word in words]
distinct_words = sorted(set(words))
print(", ".join(distinct_words))
```

```
#15 Write a Python function to create an HTML string with tags around the word(s).
def add_tags(tag, word):
    return f"<{tag}>{word}</{tag}>"

print(add_tags('i', 'Python'))
print(add_tags('b', 'Python Tutorial'))
```

```
#16 Write a Python function to insert a string in the middle of a string.
def insert_string_middle(main_string, string_to_insert):
    middle_index = len(main_string) // 2
    return main_string[:middle_index] + string_to_insert + main_string[middle_index:]

print(insert_string_middle('[[]]<<>>', 'Python'))
print(insert_string_middle('{{}}', 'PHP'))
```

```
#17 Write a Python function to get a string made of 4 copies of the last two characters of a specified string (length must be at least 2).
def insert_end(string):
    if len(string) < 2:
        return "String length must be at least 2"
    else:
        return string[-2:] * 4

print(insert_end('Python'))
print(insert_end('Exercises'))
```

```
#18 Write a Python function to get a string made of the first three characters of a specified string. If the length of the string is less than 3
def first_three(string):
    if len(string) < 3:
        return string
    else:
        return string[:3]

print(first_three('ipy'))
print(first_three('python'))
```

```
#19 Write a Python program to get the last part of a string before a specified character
def get_last_part(string, char):
    index = string.rfind(char)
    if index != -1:
        return string[:index]
    else:
        return string

url1 = "https://www.w3resource.com/python-exercises"
url2 = "https://www.w3resource.com/python"
char = "/"
print(get_last_part(url1, char))
print(get_last_part(url2, char))
```

```
#20 Write a Python function to reverse a string if its length is a multiple of 4.
def reverse_string_if_multiple_of_4(string):
    if len(string) % 4 == 0:
        return string[::-1]
    else:
        return string

print(reverse_string_if_multiple_of_4("abcd"))
print(reverse_string_if_multiple_of_4("python"))
```
