```python
#1 Write a Python script to sort (ascending and descending) a dictionary by value.
def sort_dictionary_by_value(dictionary, ascending=True):
    sorted_dict = dict(sorted(dictionary.items(), key=lambda item: item[1], reverse=not ascending))
    return sorted_dict

my_dict = {'a': 5, 'b': 2, 'c': 8, 'd': 1}
ascending_sorted_dict = sort_dictionary_by_value(my_dict)
descending_sorted_dict = sort_dictionary_by_value(my_dict, ascending=False)

print("Ascending sorted dictionary by value:", ascending_sorted_dict)
print("Descending sorted dictionary by value:", descending_sorted_dict)
```

```python
#2 Write a Python script to add a key to a dictionary.
def add_key_to_dictionary(dictionary, key, value):
    dictionary[key] = value
    return dictionary

sample_dict = {0: 10, 1: 20}
new_key = 2
new_value = 30

updated_dict = add_key_to_dictionary(sample_dict, new_key, new_value)
print("Updated dictionary:", updated_dict)
```

```python
#3 Write a Python script to concatenate the following dictionaries to create a new one.
def concatenate_dictionaries(*dictionaries):
    concatenated_dict = {}
    for dictionary in dictionaries:
        concatenated_dict.update(dictionary)
    return concatenated_dict

dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

result_dict = concatenate_dictionaries(dic1, dic2, dic3)
print("Concatenated dictionary:", result_dict)
```

```python
#4 Write a Python script to check whether a given key already exists in a dictionary.
def check_key_existence(dictionary, key):
    return key in dictionary

sample_dict = {'a': 1, 'b': 2, 'c': 3}
key_to_check = 'b'

if check_key_existence(sample_dict, key_to_check):
    print(f"The key '{key_to_check}' exists in the dictionary.")
else:
    print(f"The key '{key_to_check}' does not exist in the dictionary.")
```

```python
#5 Write a Python program to iterate over dictionaries using for loops
my_dict = {'a': 1, 'b': 2, 'c': 3}

print("Iterating over keys:")
for key in my_dict:
    print("Key:", key)

print("\nIterating over values:")
for value in my_dict.values():
    print("Value:", value)

print("\nIterating over items (key-value pairs):")
for key, value in my_dict.items():
    print("Key:", key, "Value:", value)
```

```python
#6  Write a Python script to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*x).
def generate_squared_dict(n):
    squared_dict = {x: x*x for x in range(1, n+1)}
    return squared_dict

n = 5
result_dict = generate_squared_dict(n)
print("Generated dictionary:", result_dict)
```

```python
#8 Write a Python script to merge two Python dictionaries.
def merge_dictionaries(dict1, dict2):
    merged_dict = {**dict1, **dict2}
    return merged_dict

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

merged_dict = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", merged_dict)
```

```python
#9  Write a Python program to iterate over dictionaries using for loops.
my_dict = {'a': 1, 'b': 2, 'c': 3}

print("Iterating over keys and values:")
for key in my_dict:
    print("Key:", key, "Value:", my_dict[key])

print("\nIterating over keys only:")
for key in my_dict.keys():
    print("Key:", key)

print("\nIterating over values only:")
for value in my_dict.values():
    print("Value:", value)

print("\nIterating over items (key-value pairs):")
for item in my_dict.items():
    print("Key:", item[0], "Value:", item[1])
```

```python
#10 Write a Python program to sum all the items in a dictionary.
def sum_dictionary_values(dictionary):
    return sum(dictionary.values())

my_dict = {'a': 10, 'b': 20, 'c': 30}
total_sum = sum_dictionary_values(my_dict)
print("Sum of all items in the dictionary:", total_sum)
```

```python
#11 Write a Python program to multiply all the items in a dictionary.
def multiply_dictionary_values(dictionary):
    result = 1
    for value in dictionary.values():
        result *= value
    return result

my_dict = {'a': 2, 'b': 3, 'c': 4}
total_product = multiply_dictionary_values(my_dict)
print("Product of all items in the dictionary:", total_product)
```

```python
#12 Write a Python program to remove a key from a dictionary.
def remove_key_from_dictionary(dictionary, key):
    if key in dictionary:
        del dictionary[key]
    return dictionary

my_dict = {'a': 1, 'b': 2, 'c': 3}
key_to_remove = 'b'

updated_dict = remove_key_from_dictionary(my_dict, key_to_remove)
print("Dictionary after removing key:", updated_dict)
```

```python
#13 Write a Python program to map two lists into a dictionary.
def lists_to_dictionary(keys, values):
    return dict(zip(keys, values))

keys = ['a', 'b', 'c']
values = [1, 2, 3]

result_dict = lists_to_dictionary(keys, values)
print("Mapped dictionary:", result_dict)
```

```python
#14 Write a Python program to sort a given dictionary by key.
def sort_dictionary_by_key(dictionary):
    sorted_dict = dict(sorted(dictionary.items()))
    return sorted_dict

my_dict = {'b': 2, 'a': 1, 'c': 3}

sorted_dict = sort_dictionary_by_key(my_dict)
print("Sorted dictionary by key:", sorted_dict)
```

```python
#15 Write a Python program to get the maximum and minimum values of a dictionary.
def max_min_values(dictionary):
    if len(dictionary) == 0:
        return None, None
    max_value = max(dictionary.values())
    min_value = min(dictionary.values())
    return max_value, min_value

my_dict = {'a': 10, 'b': 20, 'c': 5}

max_val, min_val = max_min_values(my_dict)
print("Maximum value in the dictionary:", max_val)
print("Minimum value in the dictionary:", min_val)
```

```python
#16 Write a Python program to get a dictionary from an object's fields.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Alice", 30)

person_dict = vars(person)

print("Dictionary from object's fields:", person_dict)
```

```python
#17 Write a Python program to remove duplicates from the dictionary.
def remove_duplicates(dictionary):
    unique_values = list(set(dictionary.values()))
    unique_dict = {key: value for key, value in dictionary.items() if value in unique_values}
    return unique_dict

my_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3, 'e': 2}

unique_dict = remove_duplicates(my_dict)
print("Dictionary after removing duplicates:", unique_dict)
```

```python
#18  Write a Python program to check if a dictionary is empty or not.
def is_empty(dictionary):
    return len(dictionary) == 0

empty_dict = {}
non_empty_dict = {'a': 1, 'b': 2}

print("Is the empty_dict empty?", is_empty(empty_dict))
print("Is the non_empty_dict empty?", is_empty(non_empty_dict))
```

```python
#19 Write a Python program to combine two dictionary by adding values for common keys.
from collections import Counter

def combine_dicts(d1, d2):
    combined_dict = Counter(d1)
    combined_dict.update(d2)
    return combined_dict

d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

result_dict = combine_dicts(d1, d2)
print("Combined dictionary:", result_dict)
```

```python
#20 Write a Python program to print all distinct values in a dictionary.
data = [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"}, {"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]

unique_values = set()
for d in data:
    unique_values |= set(d.values())

print("Unique Values:", unique_values)
```
